**Ejercicio 1. Agenda telefónica simple**

```python
# Función que recibe tres nombres y devuelve la lista
def crear_agenda(nombre1, nombre2, nombre3):
    agenda = [nombre1, nombre2, nombre3]  # Crear la lista con los tres nombres
    return agenda  # Devolver la lista completa

# Programa principal
n1 = input("Ingrese el primer nombre a llamar: ")
n2 = input("Ingrese el segundo nombre a llamar: ")
n3 = input("Ingrese el tercer nombre a llamar: ")

agenda_telefonica = crear_agenda(n1, n2, n3)
print("La primera persona que debes llamar es:", agenda_telefonica[0])  # Mostrar el primer elemento
```

**Ejercicio 2. Verificación en la lista de compras**

```python
# Función que verifica si el tercer producto es arroz
def verificar_producto(p1, p2, p3):
    compras = [p1, p2, p3]  # Crear la lista de productos
    if compras[2].lower() == "arroz":  # Comparar con "arroz", sin importar mayúsculas
        return "Ya tienes arroz en la lista"
    else:
        return "No tienes arroz en la tercera posición"

# Programa principal
prod1 = input("Ingrese el primer producto: ")
prod2 = input("Ingrese el segundo producto: ")
prod3 = input("Ingrese el tercer producto: ")

mensaje = verificar_producto(prod1, prod2, prod3)
print(mensaje)
```

**Ejercicio 3. Horario de clases del estudiante**

```python
# Función que devuelve el día central
def obtener_dia_central(dia1, dia2, dia3):
    clases = [dia1, dia2, dia3]  # Crear la lista con los días
    return clases[1]  # Retornar el segundo día (índice 1)

# Programa principal
dia_central = obtener_dia_central("lunes", "miércoles", "viernes")
print("El día central de clases es:", dia_central)
```

**Ejercicio 4. Registro parcial de asistencia**

```python
# Función que muestra los dos primeros estudiantes presentes
def asistencia_estudiantes(e1, e2, e3, e4, e5):
    presentes = [e1, e2, e3, e4, e5]  # Crear la lista con los nombres
    return presentes[0], presentes[1]  # Retornar los dos primeros

# Programa principal
nombre1 = input("Estudiante 1 presente: ")
nombre2 = input("Estudiante 2 presente: ")
nombre3 = input("Estudiante 3 presente: ")
nombre4 = input("Estudiante 4 presente: ")
nombre5 = input("Estudiante 5 presente: ")

primero, segundo = asistencia_estudiantes(nombre1, nombre2, nombre3, nombre4, nombre5)
print("Primeros estudiantes registrados:", primero, "y", segundo)
```

**Ejercicio 5. Verificación de temperatura**

```python
# Función que verifica si el martes hizo más de 30 grados
def revisar_temperatura(lunes, martes, miercoles):
    temperaturas = [lunes, martes, miercoles]  # Crear la lista con las temperaturas
    if temperaturas[1] > 30:  # Comparar temperatura del martes (índice 1)
        return "El martes hizo calor"
    else:
        return "El martes no superó los 30°C"

# Programa principal
t1 = float(input("Temperatura máxima del lunes: "))
t2 = float(input("Temperatura máxima del martes: "))
t3 = float(input("Temperatura máxima del miércoles: "))

resultado = revisar_temperatura(t1, t2, t3)
print(resultado)
```

**Ejercicio 6. Clasificación del nivel de satisfacción ciudadana**

```python
# Función que evalúa la satisfacción ciudadana según la nota del miércoles
def evaluar_satisfaccion(miercoles):
    if miercoles <= 2:
        return "Insatisfacción ciudadana"
    elif miercoles == 3:
        return "Satisfacción media"
    else:
        return "Alta satisfacción ciudadana"

# Programa principal
try:
    # Entrada de datos
    n1 = int(input("Ingrese la calificación del lunes (1 a 5): "))
    n2 = int(input("Ingrese la calificación del martes (1 a 5): "))
    n3 = int(input("Ingrese la calificación del miércoles (1 a 5): "))

    # Validación de rangos antes de enviar a la función
    if n1 < 1 or n1 > 5 or n2 < 1 or n2 > 5 or n3 < 1 or n3 > 5:
        print("Error: Las calificaciones deben estar entre 1 y 5.")
    else:
        # Si todo es válido, se llama a la función
        resultado = evaluar_satisfaccion(n3)  # Solo se evalúa el miércoles
        print("Resultado:", resultado)

except ValueError:
    print("Error: Solo se permiten números enteros entre 1 y 5.")
```
