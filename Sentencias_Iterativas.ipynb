**SENTENCIAS ITERATIVAS**

Son aquellas que permiten **ejecutar una o varias instrucciones varias veces,** es decir, **repetir un bloque** de código **mientras se cumpla una condición (sea verdadera)** o **mientras se recorren elementos de una estructura de datos** (como listas,tuplas, diccionarios).

En Python, se tienen **dos tipos de sentencias iterativas**: While y For:

repetititivo.png

**¿Por qué se necesitan?**

*   *Automatizar:* son útiles para automatizar tareas repetitivas, por ejemplo: calcular promedios de muchos elementos (estudiantes, temperaturas, etc).
*  *Procesamiento:* permiten procesar datos uno por uno, es decir, ejecutar una acción específica para cada dato, por ejemplo, imprimir, transformar, sumar

*  *Explorar:* facilitan la exploración de estructuras de datos (listas, tuplas, diccionarios, archivos, sets), es decir, examinar el contenido para analizar, validar o filtrar, para buscar elementos de acuerdo con una condición.




**Conceptos Básicos**


**Bucle (loop)**

Es una estructura que permite repetir automáticamente un bloque de código mientras se cumpla una condición o se recorra una secuencia de datos.



```python
for i in range(4):
    print("Hola")
```

**Iteración**

Cada repetición individual del bucle se llama iteración. El bucle for se ejecuta 4 veces, una por cada valor de i.

En el ejemplo anterior, se realizan cuatro iteraciones (el mensaje "Hola" se imprime 4 veces).

**Contador**

Es una variable que lleva la cuenta de cuántas veces ha ocurrido algo, o que sirve para controlar la repetición de un bucle.

```python
contador=1
while contador <= 5:
    print("Contador:", contador)
    contador += 1
```

```python
#Contar cuántos productos hay
productos = ["A", "B", "C", "D"]
contador = 0

for p in productos:
    contador += 1  # aumenta de uno en uno contandor =contador + 1

print("Cantidad de productos:", contador)
```

**Acumulador**

Es una variable que va sumando o acumulando valores dentro de un bucle.

Ejemplo: Sumar los números del 1 al 5

```python
suma = 0
for i in range(1, 6):#La función range() en Python incluye el primer número (inicio), pero excluye el último (fin).
    suma += i
    print("Iteración", i, "Suma:", suma)
print("Suma total:", suma)
```

```python
#Acumular ventas
ventas = [150, 200, 50, 400]
acumulador = 0

for v in ventas:
    acumulador += v  # suma el valor de cada venta acumulador= acumulador + 1
    print("Venta:", v, "Acumulador:", acumulador)
print("Total de ventas:", acumulador)
```

**Condición de parada**

Es una condición que indica cuándo debe detenerse el bucle.

En un while, si la condición nunca se vuelve falsa, el bucle se repite infinitamente.

```python
x = 1
while x <= 5: #Mientras x sea menor o igual a 5, haga print, e incremente x
  print(x)
  x += 1  # Condición de parada. Si no se incrementa, el bucle sería infinito, x=x+1
```

**Recorrer una secuencia**

Significa acceder a cada elemento de una colección (como lista, tupla, diccionario, sets) uno por uno, usando for.

```python
Personas = ["Ana", "Luis", "Carlos"]
"""nombre es una variable temporal que,en cada vuelta del bucle, toma uno de los valores dentro de la lista nombres.
    nombres es el iterable (lista, tupla, diccionario)"""
for x in Personas: #x es el iterador, Personas es el iterable (lista, tupla, diccionario, sets)
    print("Hola", x)
```

**LA SENTENCIA WHILE**

***¿Qué hace?***

Ejecuta un bloque de código mientras una condición sea verdadera.

***¿Cuándo se utiliza?***

Se usa en situaciones donde no se sabe con exactitud cuántas veces debe ejecutarse el bloque de código.

***Sintaxis:***

image.png

Donde:

**Condición**: expresión lógica que se verifica al comenzar cada ciclo (iteración) del bucle.

**Código a ejecutar:** conjunto de instrucciones dentro del bucle, que se ejecutarán de forma repetida siempre que la condición continúe siendo verdadera.

**Fin del bucle:** el bucle while se detiene en el momento en que la condición resulta falsa.

**Ejemplo 1. Secuencias de números del 0 al 10 al cuadrado**

```python
w = 0
while w <= 10: #La variable de control es i
	print(w**2)
	w = w + 1 #w += 1, condición de parada
```

**BUCLES INFINITOS**

Si la **condición en un ciclo while permanece siempre verdadera**, el programa entrará en una repetición constante que nunca finaliza, lo que se conoce como un ***bucle infinito.***

**¿Cómo evitarlo?**

Manejando adecuadamente las variables que intervienen en la condición del ciclo, conocidas como **variables de control.** Estas variables deben ser definidas e inicializadas antes de comenzar el ciclo while, y su valor debe modificarse dentro del cuerpo del ciclo. De esta forma, se garantiza que eventualmente la condición evaluada se vuelva falsa, permitiendo que el ciclo finalice correctamente.



**break**

Es allí, cuando la instrucción break cumple su rol. ¿Cómo? Se utiliza para **interrumpir **la ejecución de un bucle, ya sea while o for, antes de que este finalice de forma natural.
Cuando el programa encuentra esta sentencia dentro del bloque repetitivo, **el ciclo se detiene de inmediato**, sin esperar a que se cumpla la condición original.





```python
age = 25
while age > 18: # Bucle infinito. La condición siempre es verdadera
    print('Puede votar')
```

```python
age = 25
while age > 18:
    print("Puede votar")
    break  # Interrumpe el ciclo inmediatamente
```

**Controlando bucles infinitos con break en Python**

```python
# Número secreto predefinido
secreto = 7

while True:# Bucle infinito. No hay una condición para detenerse
    intento = int(input("Adivina el número (entre 1 y 10): "))

    if intento == secreto: #Condición
        print("¡Correcto! Has adivinado el número.")
        break  # Salida controlada del bucle. Cuando se ingresa el número correcto, permite salir, de lo contrario como está While True, nunca se detiene
    print("Incorrecto. Intenta de nuevo.")
```

**WHILE Y ELSE**

En Python, es posible añadir una cláusula else a un bucle while. Esta se ejecuta únicamente cuando la condición del ciclo deja de cumplirse de manera natural, sin interrupciones.

```python
saldo = 1000  # Saldo inicial en la cuenta
retiro = 200  # Monto fijo para cada retiro

while saldo >= retiro:
    saldo -= retiro #saldo=saldo-retiro
    print(f"Retiro exitoso. Saldo restante: ${saldo}")
else:
    print("No hay saldo suficiente para más retiros.")
```

```python
alumnos = {} #Se crea un diccionario vacío para guardar nombres y notas

while True:# No sabemos cuántos alumnos son
    nombre = input("Ingrese el nombre del alumno (o 'fin' para terminar): ")
    if nombre.lower() == 'fin': #Al ingresar la palabra fin, sale
    #el while True: no tiene una condición para detenerse por sí solo.
    #Sin el break, el bucle seguiría ejecutándose aunque el usuario escriba 'fin'.
       break #sale del bucle principal

    #Código para las notas de los alumnos
    notas = [] # Lista vacía para almacenar las 3 notas del alumno
    contador = 1 # Contador para saber cuántas notas se han ingresado

    while contador <= 3:  # Bucle interno: se repite hasta ingresar 3 notas
        try:
            nota = float(input(f"Ingrese la nota {contador} de {nombre} (1.0 a 7.0): "))# Solicita la nota
            if 1.0 <= nota <= 7.0:
                notas.append(nota)  # Agrega la nota válida a la lista
                contador += 1 # Aumenta el contador para la siguiente nota
            else:
                print("La nota debe estar entre 1.0 y 7.0.") # Mensaje si la nota está fuera de rango
        except ValueError: # Si el usuario escribe algo que no es número
            print("Entrada inválida. Debe ingresar un número.")

    alumnos[nombre] = notas # Se almacena el nombre del alumno y su lista de notas en el diccionario

# Mostrar resultados
print("\nResumen de alumnos:")
for alumno, notas in alumnos.items(): # Recorre el diccionario: clave = nombre, valor = lista de notas
    promedio = sum(notas) / len(notas) # Calcula el promedio de las notas
    print(f"{alumno}: Notas = {notas} | Promedio = {round(promedio, 2)}") # Muestra notas y promedio
```

**LA SENTENCIA FOR**

**¿Qué hace?**

El ciclo **for** en Python se utiliza para recorrer elementos de una secuencia (como listas, cadenas, diccionarios, tuplas, rangos.), ejecutando un bloque de código una vez por cada elemento de la secuencia.

**¿Cuándo se utiliza?**

Se usa en situaciones donde se conoce con exactitud cuántas veces debe ejecutarse el bloque de código.

**Sintaxis:**

for.png

***Conceptos***

**Iterables:** son estructuras que pueden recorrerse elemento por elemento, es decir, permiten acceder secuencialmente a sus datos. En Python, un ejemplo es una lista en Python, que se puede consultar mediante índices como lista[1], lo cual significa que es un objeto iterable.

**Iteradores:** son objetos que permiten avanzar por un iterable. Mantienen una posición interna y proporcionan un mecanismo, como el método next() para acceder al siguiente valor en la secuencia cada vez que se les llama.
El iterador, es la variable de control dentro del for.

**Ejemplo 3. Eliminar materias repetidas**

```python
materias = ["Matemáticas", "Historia", "Biología", "Historia", "Matemáticas", "Química"]
materias_sin_repetir = []

for materia in materias:
    if materia not in materias_sin_repetir:
        materias_sin_repetir.append(materia)

print("Materias sin repetir:", materias_sin_repetir)
```

```python
materias = ["Matemáticas", "Historia", "Biología", "Historia", "Matemáticas", "Química"]

for materia in materias[:]:  # Recorremos una copia de la lista
 # Para cada materia contamos cuántas veces aparece en la lista la materia
 # Mientras esa cantidad sea mayor que 1 significa que hay duplicados
    while materias.count(materia) > 1:
        materias.remove(materia) #Elimina la primera aparición de esa materia en la lista
        #while repite esta eliminación hasta que solo quede una única aparición (cuando count(materia) == 1).

print("Materias sin repetir:", materias)
```

**Ejemplo 5. Calificaciones y estado del alumno**

Clasifica alumnos según su nota usando sentencias condicionales (if-elif-else), estructura de datos (diccionario) y sentencias iterativas (for)

```python
alumnos = {
    "Ana": 6.5,
    "Luis": 4.8,
    "Marta": 3.9,
    "Carlos": 5.7
}

for alumno, nota in alumnos.items():
    if nota >= 5.5:
        estado = "Aprobado"
    elif 4.0 <= nota < 5.5:
        estado = "Recuperación"
    else:
        estado = "Reprobado"
    print(f"{alumno}: {nota} puntos - {estado}")
```

Creando el diccionario.

```python
# Crear un diccionario vacío para almacenar alumnos y sus notas
alumnos = {}

# Número de alumnos a ingresar
n = int(input("¿Cuántos alumnos quieres ingresar? "))

# Llenar el diccionario con datos ingresados por el usuario
for _ in range(n): #_ convención para indicar que no se necesita usar la variable iterador dentro del bucle
    nombre = input("Ingrese el nombre del alumno: ")
    while True:
        try:
            nota = float(input(f"Ingrese la nota de {nombre} (1.0 a 7.0): "))
            if 1.0 <= nota <= 7.0:
                break #Nota válida salir del bucle
            else:
                print("La nota debe estar entre 1.0 y 7.0. Intente de nuevo.")
        except ValueError:
            print("Entrada inválida. Por favor ingrese un número decimal.")
    alumnos[nombre] = nota

# Evaluar y mostrar el estado de cada alumno
for alumno, nota in alumnos.items():
    if nota >= 5.5:
        estado = "Aprobado"
    elif 4.0 <= nota < 5.5:
        estado = "Recuperación"
    else:
        estado = "Reprobado"
    print(f"{alumno}: {nota} puntos - {estado}")
```

**LA FUNCIÓN RANGE**

La función range() es una herramienta incorporada en Python que **genera una secuencia numérica de enteros**, útil para repetir acciones de forma controlada, como sucede en otros lenguajes de programación. Esta función puede **recibir** **hasta tres parámetros: el valor inicial, el valor final y el incremento**, aunque solo el segundo es obligatorio.

Su estructura es:

RANGE.png

Donde:

*   Inicio: define el número desde el que comienza la secuencia (si no se indica, comienza en 0).
*   Fin: determina hasta dónde llega la secuencia, aunque este valor no se incluye.
*  Paso: establece el intervalo entre los números generados (por defecto es 1).

El resultado que devuelve range() es un objeto especial de tipo range, que puede transformarse fácilmente en listas o tuplas si se necesita visualizar o manipular sus valores directamente.

*Importante:* no tiene sentido usar range() con un while, porque while funciona con condiciones lógicas, no con iteradores.

```python
print(range(4))
print(list(range(4)))        # Convierte rango a lista
```

```python
print(list(range(2, 5)))
```

**ENUNCIADO DEL PROBLEMA**

*Análisis de satisfacción de clientes en una tienda online*

Una tienda online desea analizar la satisfacción de sus clientes basada en las puntuaciones que reciben tras cada compra. Los datos están compuestos por una lista de clientes, donde cada cliente tiene un nombre y una lista de puntuaciones (del 1 al 5) que ha dado en diferentes compras.

Se pide un programa que:

1. Reciba la información de varios clientes y sus puntuaciones.

2. Calcule el promedio de satisfacción para cada cliente.

3. Clasifique a cada cliente según su promedio:

    Satisfecho si el promedio es mayor o igual a 4.0

    Neutral si el promedio está entre 3.0 y 3.9

    Insatisfecho si es menor a 3.0

7. Identifique y muestre el cliente con la mayor satisfacción.

8. Permita mostrar un reporte con todos los clientes, sus promedios y clasificaciones.

```python
# Función para calcular el promedio
def calcular_promedio(puntuaciones):
    if len(puntuaciones) == 0:
        return 0  # Evitar división por cero
    return sum(puntuaciones) / len(puntuaciones)

# Función para clasificar la satisfacción
def clasificar_satisfaccion(promedio):
    if promedio >= 4.0:
        return "Satisfecho"
    elif 3.0 <= promedio < 4.0:
        return "Neutral"
    else:
        return "Insatisfecho"

# Función principal Recibe información de varios clientes y sus puntuaciones
def main():
    clientes = {}      # Diccionario para almacenar puntuaciones por cliente
    resultados = {}    # Diccionario con promedio y clasificación por cliente

    # Ingresar número de clientes
    n = int(input("¿Cuántos clientes desea registrar? "))

    # Recolectar datos por cliente
    for _ in range(n):
        nombre = input("\nIngrese el nombre del cliente: ")
        m = int(input(f"Ingrese el número de puntuaciones para {nombre}: "))
        puntuaciones = []  # Lista de puntuaciones por cliente

        for i in range(m):
            while True:
                try:
                    p = float(input(f"Ingrese puntuación {i+1} (1 a 5): "))
                    if 1 <= p <= 5:
                        puntuaciones.append(p)
                        break # Si la nota es válida
                    else:
                        print("La puntuación debe estar entre 1 y 5. Intente de nuevo.")
                except ValueError:
                    print("Entrada inválida. Ingrese un número decimal.")

        clientes[nombre] = puntuaciones  # Asociar cliente con sus puntuaciones

    # Calcular promedio y clasificación por cliente
    for cliente, punt_list in clientes.items():
        promedio = calcular_promedio(punt_list)
        clasificacion = clasificar_satisfaccion(promedio)
        resultados[cliente] = {
            "promedio": promedio,
            "clasificacion": clasificacion
        }

    # Determinar cliente con mayor promedio
    #Sintaxis lambda argumentos: expresión

#Encuentra el nombre del cliente (cliente_top) que tiene el promedio más alto de satisfacción
#dentro del diccionario resultados
#Por defecto compara por claves, por nombres.
#Por eso, le damos una función personalizada para decirle cómo comparar
#Función anónima (lambda) que toma cada clave c (por ejemplo, "Ana", "Luis", "Carla") y devuelve su promedio4

    cliente_top = max(resultados, key=lambda c: resultados[c]["promedio"])
    promedio_top = resultados[cliente_top]["promedio"]

    # Mostrar reporte
    print("\n--- Reporte de satisfacción de clientes ---")

  #Recorremos el diccionario resultados
  #cliente toma el nombre del cliente,
  #datos toma el diccionario interno conpromedio y clasificación

  #for clave, valor in diccionario.items():
  #cliente y datos son variables temporales que reciben cada par del diccionario (clave, valor)
  #No se deben declarar, se usan dentro del ciclo como variables iteradoras.

    for cliente, datos in resultados.items():
        print(f"{cliente}: Promedio = {datos['promedio']:.2f}, Clasificación = {datos['clasificacion']}")
        #Salida print:
        #Ana: Promedio = 4.30, Clasificación = Satisfecho
        #Luis: Promedio = 3.50, Clasificación = Neutral

    print(f"\nCliente con mayor satisfacción: {cliente_top} (Promedio: {promedio_top:.2f})")

# 4. Ejecutar programa principal
if __name__ == "__main__": #Buenas prácticas en Python para ejecutar el programa principal
    main()
```
