**EJERCICIO 1.**

```python
def mostrar_producto(producto):
    nombre, precio, cantidad = producto
    print(f"\nProducto: {nombre}")
    print(f"Precio: ${precio:.2f}")
    print(f"Cantidad en stock: {cantidad}")

    if cantidad >= 10:
        print("Stock suficiente.")
    else:
        print("Stock limitado.")

# Programa principal
try:
    nombre = input("Ingrese el nombre del producto: ").strip().capitalize()
    precio_str = input("Ingrese el precio: ").strip().replace(",", ".")
    cantidad_str = input("Ingrese la cantidad disponible: ").strip()

    if not nombre or not precio_str or not cantidad_str:
        raise ValueError("Debe ingresar todos los datos.")

    precio = float(precio_str)
    cantidad = int(cantidad_str)

    if precio < 0 or cantidad < 0:
        raise ValueError("Ni el precio ni la cantidad pueden ser negativos.")

    producto = (nombre, precio, cantidad)
    mostrar_producto(producto)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Ocurrió un error inesperado:", e)
```

**EJERCICIO 2.**

```python
# Función para analizar la ciudad
def analizar_ciudad(datos_ciudad):
    ciudad, region, temperatura = datos_ciudad
    print(f"\nCiudad: {ciudad}")
    print(f"Región: {region}")
    print(f"Temperatura promedio: {temperatura}°C")

    if temperatura < 15:
        print("Clima: Frío")
    elif temperatura < 25:
        print("Clima: Templado")
    else:
        print("Clima: Cálido")

# Programa principal
try:
    ciudad = input("Ingrese el nombre de la ciudad: ").strip().capitalize()
    region = input("Ingrese la región: ").strip().capitalize()
    temp_str = input("Ingrese la temperatura promedio (°C): ").strip().replace(",", ".")

    # Validación de campos vacíos
    if not ciudad or not region or not temp_str:
        raise ValueError("Todos los campos son obligatorios.")

    # Conversión segura
    temperatura = float(temp_str)
    if temperatura < -20 or temperatura > 50:
        raise ValueError("La temperatura ingresada no es realista.")

    # Crear la tupla y analizarla
    datos = (ciudad, region, temperatura)
    analizar_ciudad(datos)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Ocurrió un error inesperado:", e)
```
