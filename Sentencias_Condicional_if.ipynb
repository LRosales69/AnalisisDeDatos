```python
#Recordar
m = 5
n = "5"
l = 5.0
print(m == n)  # Compara valores y tipos
print(m == l)  # Compara valores y tipos

m = "hola"
n = "holita"
k = m == n
print(type(m))
print(type(n))
print(k, type(k))

#Asignar varios valores a varias variables
a, b, c = 1, 2, 3 #Múltiples asignaciones
print(a)  # 1
print(b)  # 2
print(c)  # 3

#Asignar el mismo valor a varias variables
x = y = z = 100
print(x, y, z)  # 100 100 100

#Intercambiar valores entre variables (swap)
a = 5
b = 10
a, b = b, a
print(a, b)  # 10 5 #Se intercambian los valores sin tener que usar variables temporales
```

```python
#Ejemplo 1. Evaluación de Expresiones
#Reglas de precedencia lógica NOT-AND-OR

# Valores de las variables
X = True
Y = False

# Evaluamos la expresión lógica
resultado_a = not (X and Y or X)

print("Resultado de la expresión a:", resultado_a)
```

```python
##Ejemplo 2. Semáforo
semaforo = input("Color del semáforo (rojo, amarillo, verde): ").lower() #No hay conversión, porque el color del semáforo ingresa es una cadena
if semaforo == "verde":
  print ("Cruzar la calle" )
#¿Qué pasa si el usuario ingresa un color distinto de verde?
```

```python
##Ejemplo 3. Validar mayoría de edad
edad = int(input("¿Cuántos años tienes?")) #Conversión de datos

if edad >= 18:
    print("Eres mayor de edad.")
#¿Qué pasa si la persona no es mayor de edad?
```

```python
##Ejemplo 4. Número Positivo
numero = float(input("Ingresa un número: "))
if numero > 0:
    print("El número es positivo.")
#¿Qué pasa si el número no es positivo?
```

```python
##Ejemplo 5. Verificar si un número es par

numero = int(input("Ingresa un número entero: "))

if numero % 2 == 0:
    print("El número es par.")
#¿Qué pasa si el número no es par?
```

```python
##Ejemplo 6. Verificar si una palabra contiene una letra específica
palabra = input("Escribe una palabra: ")

if "a" in palabra:
    print("La palabra contiene la letra 'a'.")
#¿Qué pasa si la palabra no contiene la letra 'a'?
```

```python
##Ejemplo 7. Compra
totalcompra = float(input("Ingresa el total de la compra: "))
importeapagar = totalcompra
if totalcompra > 100:
  tasadescuento = 10
  importedescuento = totalcompra * tasadescuento / 100
  importeapagar = totalcompra-importedescuento
  print("El importe a pagar es:", importeapagar)
```
