**Ejercicio 1. Rectángulo**

```python
# Definimos la clase Rectangulo
class Rectangulo:
    def __init__(self, ancho, alto):
        self.ancho = ancho
        self.alto = alto

    def area(self):
        # Retorna el área (ancho × alto)
        return self.ancho * self.alto

    def perimetro(self):
        # Retorna el perímetro (2 × (ancho + alto))
        return 2 * (self.ancho + self.alto)

# Creamos un rectángulo con ancho 4 y alto 6
mi_rectangulo = Rectangulo(4, 6)

# Mostramos el área y el perímetro
print("Área:", mi_rectangulo.area())
print("Perímetro:", mi_rectangulo.perimetro())
```

**Ejercicio 2. Producto con descuento**

```python
# Definimos la clase Producto
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def precio_con_descuento(self,porcentaje):
        # Calcula el descuento y lo resta del precio original
        descuento = self.precio * (porcentaje / 100)
        return self.precio - descuento

# Creamos un producto
producto1 = Producto("Audífonos", 100) #Instanciación de un objeto de la clase producto

# Calculamos el precio con 15% de descuento
precio_final = producto1.precio_con_descuento(15)

# Mostramos el resultado
print(f"Precio con 15% de descuento: ${precio_final}")
```

**Ejercicio 3. Estudiante y promedio**

```python
# Definimos la clase Estudiante
class Estudiante:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas  # Lista con 3 notas

    def promedio(self):
        # Calcula el promedio de las notas
        return sum(self.notas) / len(self.notas)

# Creamos un estudiante con tres notas
estudiante1 = Estudiante("Marta", [18, 20, 17])

# Calculamos y mostramos el promedio
print(f"Promedio de {estudiante1.nombre}: {estudiante1.promedio():.2f}")
```

**Ejercicio 4. Gestión de empleados y análisis de datos**

Una empresa desea analizar a sus empleados registrados en un archivo de datos, entre ellos a quienes ocupan el cargo de "Analista de Datos". Para facilitar esta tarea, se te solicita desarrollar una clase llamada AnalistaDatos que permita trabajar con un conjunto de empleados almacenado en un DataFrame de Pandas, el cual incluye la siguiente información: nombre, edad, salario y cargo.

La clase deberá permitir:

1. Filtrar y mostrar solo a los empleados cuyo cargo sea "Analista de Datos".

2. Calcular el salario promedio exclusivamente de los analistas de datos.

3. Identificar al analista de datos más experimentado, es decir, el de mayor edad.

4. Se espera que el programa entregue los resultados de forma clara y utilice funciones de Pandas integradas para facilitar el análisis.

```python
import pandas as pd

class AnalistaDatos:
    def __init__(self, datos):
        # Se guarda el DataFrame completo
        self.datos = datos

    def filtrar_analistas(self):
        # Retorna solo las filas donde el cargo es "Analista de Datos"
        return self.datos[self.datos['cargo'] == 'Analista de Datos']

    def salario_promedio_analistas(self):
        # Calcula el salario promedio solo de los analistas de datos
        analistas = self.filtrar_analistas()
        return analistas['salario'].mean()

    def analista_mas_experimentado(self):
        # El empleado más viejo entre los analistas
        analistas = self.filtrar_analistas()
        idx = analistas['edad'].idxmax() #Asigna la fila del analista con edad máxima
        return analistas.loc[idx, 'nombre'] #Accedo a la fila idx y columna nombre

# Datos de ejemplo
datos_empleados = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Carlos', 'Marta'],
    'edad': [28, 35, 22, 30],
    'salario': [3500, 4200, 3000, 3800],
    'cargo': ['Analista de Datos', 'Gerente', 'Analista de Datos', 'Contadora']
})

# Crear objeto AnalistaDatos con todos los empleados
analista = AnalistaDatos(datos_empleados)

# Resultados
print("Analistas de datos:\n", analista.filtrar_analistas())
print("\nSalario promedio de analistas:", analista.salario_promedio_analistas())
print("Analista más experimentado:", analista.analista_mas_experimentado())
```
