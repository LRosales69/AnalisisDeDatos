FUNCIONES PERSONALIZADAS

La siguiente función determina si un número ingresado por el usuario es múltiplo de 3. ¿Cuál es la lógica a seguir? 1. Solicitar un número al usuario. 2. Verificar si el número módulo 3 (% 3) es igual a 0, y 3.Mostrar el resultado.

1. FUNCIÓN MÚLTIPLO

```python
#Función múltiplo
#Definición de la función
def es_multiplo(numero):
    if numero % 3 == 0: #Evaluó la condición, si es verdadera retorna TRUE
        return True
    else:
        return False  #Si no se cumple el if, retorna FALSE

#¿Cómo utilizar la función?
#Programa Principal
num = int(input("Ingresa un número: ")) #Entrada de datos: Solicito al usuario que ingrese un número.
if es_multiplo(num): #Hago el llamado de la función multiplo_de_3 que ya fue definida (invocar la función)
    print("Es múltiplo de 3") # Si la condición del if es verdadera (se cumple), devuelve este mensaje
else:
    print("No es múltiplo de 3") #Si la condición del if es falsa (no se cumple), devuelve este mensaje
```

**EXPLICACIÓN DE LA FUNCIÓN ES_MÚLTIPLO**

**Definición de la función (def es_multiplo(numero):)**

*def *es una palabra clave que indica que se va a definir una función.

*multiplo_de_3 *es el nombre de la función (puede tener cualquier nombre, pero debe ser claro y sin espacios).

*(numero)* es el parámetro de entrada.
Es decir, la función necesita que le pases un valor (en este caso, un número) para poder trabajar con él.

*Los dos puntos :* indican el inicio del bloque de código que pertenece a la función.

**Condicional if (if numero % 3 == 0:):**

Aquí se verifica una condición: si el número ingresado es divisible por 3.

numero % 3 usa el operador módulo %, que devuelve el residuo de dividir numero entre 3.

Si numero % 3 == 0, significa que no hay residuo y por lo tanto el número es múltiplo de 3.

Si la condición es verdadera, se ejecuta el bloque indentado (con espacio al inicio) que está debajo.

**Retorno si la condición se cumple (return True):**

*return* es una instrucción que devuelve un valor al lugar donde se llamó la función.

En este caso, si el número es múltiplo de 3, devuelve True (Verdadero).

**Alternativa else:**

Si la condición del if NO SE CUMPLE, se ejecuta el bloque dentro del else.
Es decir, si el número NO es múltiplo de 3, se pasará al siguiente bloque.

**Retorno si la condición NO se cumple (return False):**

Si el número ingresado NO ES múltiplo de 3, la función devuelve False.



2. FUNCIÓN RAÍZ CUADRADA

```python
#función raíz cuadrada
def raiz_cuadrada (numero):
  if numero < 0:
     return "Error: El número debe ser positivo"
  else:
     return numero **0.5

#Como usar la función?
#programa principal
num = float(input("Ingresa un número para calcular su raíz cuadrada: "))
resultado = raiz_cuadrada (num)
print("Resultado:",resultado)
```

**EXPLICACIÓN DE LA FUNCIÓN RAIZ CUADRADA**
# Definición de la función
En primer lugar, tenemos que **definir** una **función llamada raiz_cuadrada.** Esta función tiene un **solo parámetro: numero**, que será el valor del que se desea calcular la raíz cuadrada. La palabra clave **def** indica que estamos **definiendo una función, seguida del nombre de la función** y los paréntesis con el **parámetro.**

# Sentencias Condicionales
Dentro de la función se utiliza una **estructura condicional if-else**. El primer paso es comprobar si el número recibido como parámetro es negativo. Esto se hace con la línea **if numero < 0:**. Si la condición se cumple (es decir, si el número es menor que cero), la función devuelve un mensaje de error como texto: "Error: El número debe ser positivo". Esto es porque en matemáticas reales, no se puede calcular la raíz cuadrada de un número negativo sin entrar en el ámbito de los números complejos, y el objetivo aquí es mantenerlo simple y entendible en este nivel básico.

**Si el número no es negativo**, entonces se ejecuta la parte del else. En esta rama del condicional, se calcula la raíz cuadrada utilizando la operación numero ** 0.5 ** **texto en negrita**. Esta notación en Python significa "elevar el número a la potencia de 0.5", que es equivalente a calcular su raíz cuadrada. Luego se devuelve ese resultado.

# Programa Principal
Después de la definición de la función, viene el programa principal, que es el lugar donde se usa esa función para resolver un problema concreto. Aquí se le pide al usuario que ingrese un número, utilizando la **función input()**. Como input() devuelve una cadena de texto, es necesario convertirla a número decimal usando float() para poder hacer cálculos matemáticos.

La línea **num = float(input(...))** guarda ese valor convertido en la variable num.

A continuación, se llama a la función raiz_cuadrada() pasándole como argumento el valor que ingresó el usuario. El resultado que devuelva la función (ya sea el valor de la raíz cuadrada o el mensaje de error) se almacena en la variable resultado.

Por último, se imprime el resultado en pantalla con la instrucción **print("Resultado:", resultado).** Esto mostrará ya sea el valor calculado de la raíz cuadrada o el mensaje "Error: El número debe ser positivo" si se intentó calcular la raíz de un número negativo.

4. FUNCIÓN QUE DETERMINA SI UN NÚMERO ES NEGATIVO, POSITIVO O CERO

```python
# Definición de la función
def tipo_de_numero(numero):
    """Determina si un número es positivo, negativo o cero."""
    if numero > 0: #Si el número es positivo, retorna Positivo
        return "Positivo"
    elif numero < 0: #Sino, si el numero es menor que cero, retorna Negativo
        return "Negativo"
    else: #Sino, retorna cero
        return "Cero"

# Programa principal
try:#Manejo de errores
    valor = float(input("Ingresa un número: "))#Entrada de datos
    resultado = tipo_de_numero(valor)#Llamado a la función
    print("El número es:", resultado)
except ValueError:#Manejo de errores
    print("Error: Debes ingresar un número válido.")#Salida
```

**EXPLICACIÓN DE LA FUNCIÓN tipo_de_numero**
# def tipo_de_numero(numero):
Primero, se define una función personalizada llamada tipo_de_numero que recibe como parámetro un número. Esta función sirve para analizar ese número y determinar si es positivo, negativo o cero. La definición inicia con la palabra clave **def**, seguida del **nombre de la función** y los paréntesis con el **parámetro** entre ellos.

# Sentencias Condicionales if-elif-else
Dentro de la función se usa una **estructura condicional if-elif-else** para evaluar el valor del número:

Si el número es mayor que cero **(if numero > 0:)**, la función devuelve la cadena de texto "Positivo".

Si no es mayor que cero pero es menor **(elif numero < 0:)**, entonces devuelve "Negativo".

Si no cumple ninguna de las condiciones anteriores, significa que el número es igual a cero, por lo tanto se ejecuta el **else **y se devuelve "Cero".

# Programa Principal
Luego, en el **programa principal**, se pide al usuario que ingrese un número. Para ello, se utiliza la función **input(),** **pero como input() devuelve siempre una cadena de texto**, es necesario **convertirla a número utilizando float**(), ya que los valores ingresados podrían ser decimales.

**Como la conversión de texto a número puede fallar si el usuario escribe algo que no es un número (por ejemplo, “hola” o “tres”),** se utiliza un **bloque try-except.** Este bloque intenta ejecutar el código dentro del try y, si ocurre un error de conversión, pasa al bloque except y muestra un mensaje de error personalizado, evitando que el programa se detenga bruscamente.

Dentro del bloque try, después de obtener el número con float(input(...)), se llama a la **función tipo_de_numero()** pasándole el número ingresado como **argumento.** El resultado que devuelve la función se guarda en la variable resultado.

Por último, se muestra el resultado al usuario usando **print(),** indicando si el número es positivo, negativo o cero.



5. FUNCIÓN QUE CALCULA EL PROMEDIO DE TRES NOTAS

```python
# Definición de la función
def calcular_promedio(n1, n2, n3):
    """Calcula el promedio de tres notas."""
    #promedio = (n1 + n2 + n3) / 3
    return (n1 + n2 + n3) / 3

# Programa principal
try:
    # Solicitar el ingreso de notas al usuario
    nota1 = float(input("Ingresa la primera nota: "))
    nota2 = float(input("Ingresa la segunda nota: "))
    nota3 = float(input("Ingresa la tercera nota: "))

    # Validación: notas estén dentro del rango (1.0 a 7.0)
    if 1.0 <= nota1 <= 7.0 and 1.0 <= nota2 <= 7.0 and 1.0 <= nota3 <= 7.0:
      """Llamado a la función calcular_promedio y se le pasan tres argumentos nota1, nota2, nota3"""
      promedio_final = calcular_promedio(nota1, nota2, nota3)
      print("Promedio final:", round(promedio_final, 2))# Muestra en pantalla el promedio de la tres notas, con dos decimales

      if promedio_final >= 4.0:# Si se cumple la condición mostrará Aprobado
            print("Resultado: APROBADO")
      else:
            print("Resultado: REPROBADO") # Si no se cumple la condición del if, mostrará Reprobado
    else:
        #Mensaje de error, si el usuario ingresa notas fuera del rango establecido
        print("Error: Las notas deben estar entre 1.0 y 7.0.")

except ValueError:
    print("Error: Debes ingresar valores numéricos.")
```
