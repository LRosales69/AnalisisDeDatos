**Análisis de temperaturas semanales**

Una empresa desea analizar las temperaturas registradas durante una semana. Se necesita:

Ingresar 7 temperaturas (una por día).

Calcular el promedio.

Contar cuántos días estuvieron por encima del promedio.



```python
# Función para calcular el promedio
def calcular_promedio(lista):
    return sum(lista) / len(lista)

# Función principal del análisis
def analizar_temperaturas():
    temperaturas = []  # Lista para almacenar los datos
    print("Ingrese la temperatura registrada en cada uno de los 7 días:")

    for i in range(1, 8):  # 7 días
        temp = float(input(f"Día {i}: "))
        temperaturas.append(temp)

    promedio = calcular_promedio(temperaturas)

    dias_mayores = 0
    for temp in temperaturas:
        if temp > promedio:
            dias_mayores += 1

    print(f"\nTemperaturas registradas: {temperaturas}")
    print(f"Promedio semanal: {promedio:.2f}°")
    print(f"Días con temperatura por encima del promedio: {dias_mayores}")

    # Programa principal
if __name__ == "__main__":
    analizar_temperaturas()
```

**Análisis de asistencia semanal**

Una escuela desea analizar la asistencia de un grupo durante la semana. Se ingresan los días en los que un estudiante asistió. El sistema debe:

Registrar asistencia de 5 días.

Calcular el porcentaje de asistencia.

Indicar si el estudiante aprobó la asistencia (mínimo 80%).

```python
# Función para calcular el porcentaje de asistencia
def calcular_porcentaje(asistencias, total_dias):
    return (asistencias / total_dias) * 100

# Función principal
def analizar_asistencia():
    total_dias = 5
    asistencias = 0

    print("Ingrese '1' si asistió o '0' si faltó en cada uno de los 5 días:")

    for i in range(1, total_dias + 1):
        while True:
            try:
                entrada = int(input(f"Día {i}: "))
                if entrada in [0, 1]:
                    asistencias += entrada
                    break
                else:
                    print("Ingrese solo 1 (asistió) o 0 (faltó).")
            except ValueError:
                print("Entrada inválida. Ingrese un número entero (0 o 1).")

    porcentaje = calcular_porcentaje(asistencias, total_dias)

    print(f"\nDías asistidos: {asistencias}")
    print(f"Porcentaje de asistencia: {porcentaje:.2f}%")
    if porcentaje >= 80:
        print("Asistencia aprobada")
    else:
        print("Asistencia insuficiente")

# Programa principal
if __name__ == "__main__": #condición especial que se usa para controlar cuándo se ejecuta un bloque de código.
    analizar_asistencia()
```

**Cálculo de ingresos diarios**

Un negocio desea registrar sus ingresos diarios durante una semana. El sistema debe:

Recibir 7 valores de ingresos.

Calcular el total semanal.

Determinar el día con el mayor ingreso.

```python
# Función para calcular el total
def calcular_total(lista):
    return sum(lista)

# Función para encontrar el mejor día
def mejor_dia(lista):
  #max(lista) encuentra el valor máximo dentro de la lista
  #lista.index busca la posición donde aparece ese valor máximo por primera vez
  #Se suma 1, para que aparezca día 1, etc. El día 1 corresponde al índice 0, día 2 al índice 1, etc.
    return lista.index(max(lista)) + 1  # +1 porque los días comienzan en 1

# Función principal
def analizar_ingresos():
    ingresos = []

    print("Ingrese los ingresos obtenidos cada día durante 7 días:")

    for i in range(1, 8):
        while True:
            try:
                ingreso = float(input(f"Día {i}: $"))
                if ingreso >= 0:
                    ingresos.append(ingreso)
                    break
                else:
                    print("El ingreso no puede ser negativo.")
            except ValueError:
                print("Entrada inválida. Ingrese un número.")

    total = calcular_total(ingresos)
    dia_max = mejor_dia(ingresos)

    print(f"\nIngresos: {ingresos}")
    print(f"Total semanal: ${total:.2f}")
    print(f"Mayor ingreso: Día {dia_max} (${ingresos[dia_max - 1]:.2f})")

# Programa principal
if __name__ == "__main__":
    analizar_ingresos()
```
