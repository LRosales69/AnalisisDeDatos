**ESTRUCTURA DE DATOS**

Son aquellas que permiten almacenar y gestionar datos en la memoria de un programa de manera organizada, para facilitar su acceso y modificación.

Las estructuras de datos básicas en Python, son: listas, tuplas, diccionarios, conjuntos (sets).

**¿POR QUÉ SE NECESITAN?**

Se necesitan porque:

*  Permiten organizar datos de manera eficiente.
*  Mejoran el rendimiento del programa.
*   Facilitan operaciones como búsqueda, inserción, eliminación y recorrido de datos.

**LISTAS**

Es una estructura de datos que permite almacenar varios elementos en un orden específico. Lo que la hace especialmente útil es que su contenido puede cambiar, es decir, se pueden agregar, eliminar o modificar valores después de haberla creado. Además, una lista puede contener distintos tipos de datos al mismo tiempo, como números, textos u otras listas, lo que la convierte en una herramienta muy flexible para manejar conjuntos de información de manera dinámica.

Características
*   Se declaran entre corchetes [ ].
*   Modificables (son mutables).
*   Pueden contener distintos tipos de datos.
*   Permiten duplicados.
*   Los elementos permanecen en el mismo orden en que fueron agregados a la lista (son ordenadas)

Estructura General

lista = [elemento1, elemento2, elemento3, ..., elementoN]

lista=list((elemento1, elemento2,....,elementoN))

**Acceso de elementos desde el final de la lista**

Suponga que se tiene la siguiente lista, llamada lista de prueba (Lista_p):

Lista_p = ['a', 'b', 'c', 'd', 'e']

índices positivos:  [0,   1,   2,   3,   4]

índices negativos: [-5,  -4,  -3,  -2,  -1]

**# ¿Qué hace Lista_p[-3:-1]?**

La instrucción Lista_p[-3:-1], extrae los elementos desde el índice -3 (inclusive) hasta el índice -1 (exclusivo). Es decir:

Lista_p[-3] → 'c'

Lista_p[-2] → 'd'

Lista_p[-1] → 'e'  ← este no se incluye

Resultado: ['c', 'd']

**¿Por qué no funciona si se coloca Lista_p[-1:-3]?**

No funciona porque el slicing en Python (tomar un parte, un segmento, un subconjunto) va hacia adelante, es decir: Python va del índice inicial al final, de izquierda a derecha, mientras el índice de inicio sea menor que el de fin.

Si se coloca Lista_p[-1:-3], le estamos indicando a Python que: empiece en -1 (valor 'e') y termina en -3 (valor 'c'). Pero, como -1 es mayor que -3, y el slicing por defecto no va hacia atrás, el resultado será:

[]  # una lista vacía

```python
#Acceso a los elementos de una lista (desde el final)
print(frutas[-3:-1])  # Desde índice 0 hasta antes del 2 (muestra elementos en posición 0 y 1)
```

**LISTAS Y CADENAS DE CARACTERES**

Se puede convertir una cadena en una lista de caracteres.
Puede dividir una cadena en una lista de caracteres. Además, puede dividir una cadena en una lista de palabras mediante el método split(). El delimitador predeterminado para el método split() es un espacio en blanco. Sin embargo, puede especificar qué caracteres usar como límites de palabras. Por ejemplo, puede usar un guion como delimitador

```python
#Utilizando la función list para dividir una lista en una cadena de caracteres
palabra_1 = "Python" # Se guarda la palabra "Python" como una cadena de texto
letras = list(palabra_1) #La función list() toma cada carácter de la cadena y lo pone como un elemento independiente en una lista.
print(letras)

#Unir
print("".join(letras))
```

```python
#Utilizando split
palabra_2 = "Python es un lenguaje de programación"
lista = palabra_2.split() # Divide la cadena por espacios en blanco (es el comportamiento por defecto)
print(lista)

#Utilizando split
palabra_3 = "Python-es-un-lenguaje-de-programación"
lista = palabra_3.split() #No funciona, no separa la cadena, porque no reconoce otro tipo de separador que no sea "espacios en blanco"
print(lista)
```

```python
#Utilizando el delimitador (-)
palabra_deli = "Python-es-un-lenguaje-de-programación"
lista = palabra_deli.split("-")
print(lista)

#Unir: "separador".join(lista)
print("-".join(lista))
```

```python
#Utilizando el delimitador (";")
palabra_deli= "Lunes;Martes;Miércoles;Jueves;Viernes"
l= palabra_deli.split(";")
print(l)

#Unir: "separador".join(nombre de la variable que almacena el resultado del split, el cual es una lista)
print(";".join(l))
```

**EJEMPLO 2. MODIFICAR UNA LISTA**

```python
# Definición de la función
#Usando append()
def agregar_animales(a1, a2):
    animales = ["perro", "gato"]
    animales.append(a1)   # Agrega el primer animal
    animales.append(a2)   # Agrega el segundo animal
    return animales       # Devuelve la lista actualizada

# Programa principal
nuevo1 = input("Ingrese un animal para agregar: ")
nuevo2 = input("Ingrese otro animal para agregar: ")

lista_actualizada = agregar_animales(nuevo1, nuevo2)
print("Lista de animales:", lista_actualizada)
```

```python
#Definición de la función
#Usando extend()
def agregar_varios(a1, a2):
    animales = ["perro", "gato"]
    animales.extend([a1, a2])  # Agrega ambos al mismo tiempo
    return animales

# Programa principal
try:
    nuevo1 = input("Ingrese un animal para agregar: ")
    nuevo2 = input("Ingrese otro animal para agregar: ")

# Validación: al menos uno debe estar lleno
    if not nuevo1 and not nuevo2:
      raise ValueError("Debe ingresar al menos un animal.")

    lista_final = agregar_varios(nuevo1, nuevo2)#llamado a la función
    print("Lista de animales:", lista_final)

except Exception as e:
    print("Ocurrió un error inesperado:", e)
```

 Y, no es lo que se busca cuando queremos agregar elementos.

**Ejemplo 3. Evaluación de tiempos de atención al cliente**

```python
from statistics import mean  # Importamos la función mean para calcular promedios

# Función que evalúa el tiempo promedio de atención
def evaluar_atencion(tiempos):
    promedio = mean(tiempos)  # Calculamos el promedio de los 3 tiempos

    # Condicional para clasificar el nivel de atención
    if promedio <= 10:
        return "Atención eficiente"
    elif promedio <= 20:
        return "Atención aceptable"
    else:
        return "Atención lenta, revisar proceso"

# Programa principal
try:
    # Solicitar los tiempos de atención
    tiempo1 = float(input("Tiempo en caja (en minutos): "))
    tiempo2 = float(input("Tiempo en asesoría (en minutos): "))
    tiempo3 = float(input("Tiempo en entrega (en minutos): "))

    # Validar que los tiempos sean positivos
    if tiempo1 < 0 or tiempo2 < 0 or tiempo3 < 0:
        print("Error: los tiempos no pueden ser negativos.")
    else:
        tiempos_cliente = [tiempo1, tiempo2, tiempo3]  # Crear la lista de tiempos
        resultado = evaluar_atencion(tiempos_cliente)  # Evaluar la atención, llamado a la función
        print("Resultado de la atención al cliente:", resultado)

except ValueError:
    print("Error: debes ingresar solo números válidos.")
```

```python
from statistics import mean  # Importar función para calcular promedio

# Función que evalúa el tiempo promedio de atención
def evaluar_atencion(tiempos):
    promedio = mean(tiempos)  # Calculamos el promedio de los tiempos

    if promedio <= 10:
        return "Atención eficiente"
    elif promedio <= 20:
        return "Atención aceptable"
    else:
        return "Atención lenta, revisar proceso"

# Programa principal
try:
    # Entrada de datos
    tiempo_caja = float(input("Tiempo en caja (en minutos): "))
    tiempo_asesoria = float(input("Tiempo en asesoría (en minutos): "))
    tiempo_entrega = float(input("Tiempo en entrega (en minutos): "))

    # Validar que los tiempos sean positivos
    if tiempo_caja < 0 or tiempo_asesoria < 0 or tiempo_entrega < 0:
        print("Error: los tiempos no pueden ser negativos.")
    else:
        tiempos = [tiempo_caja, tiempo_asesoria, tiempo_entrega]  # Crear lista

        # Evaluar la atención según el promedio
        resultado = evaluar_atencion(tiempos)
        print("\nResultado de la atención al cliente:", resultado)

        # 1. Mostrar un tiempo específico por posición
        print("\nTiempos individuales:")
        print("Tiempo en caja:", tiempos[0], "minutos")
        print("Tiempo en asesoría:", tiempos[1], "minutos")
        print("Tiempo en entrega:", tiempos[2], "minutos")

        # 2. Mostrar el tiempo mayor y menor
        print("\nTiempos extremos:")
        print("Tiempo más largo:", max(tiempos), "minutos")
        print("Tiempo más corto:", min(tiempos), "minutos")

        # 3. Mostrar todos los tiempos con sus etiquetas (sin bucles)
        etiquetas = ["Caja", "Asesoría", "Entrega"]
        print("\nResumen detallado de tiempos:")
        print(etiquetas[0] + ":", tiempos[0], "minutos")
        print(etiquetas[1] + ":", tiempos[1], "minutos")
        print(etiquetas[2] + ":", tiempos[2], "minutos")

except ValueError:
    print("Error: debes ingresar solo números válidos.")
```

```python
# Crear un diccionario base con tres claves
estudiante = {
    'nombre': 'Lucía',
    'edad': 20,
    'carrera': 'Contabilidad'
}

# 1. dict.keys(): devuelve las claves
print("Claves del diccionario:")
print(estudiante.keys())  # dict_keys(['nombre', 'edad', 'carrera'])

# 2. dict.values(): devuelve los valores
print("\nValores del diccionario:")
print(estudiante.values())  # dict_values(['Lucía', 20, 'Contabilidad'])

# 3. dict.items(): devuelve pares (clave, valor)
print("\nElementos clave-valor del diccionario:")
print(estudiante.items())  # dict_items([('nombre', 'Lucía'), ('edad', 20), ('carrera', 'Contabilidad')])

# 4. dict.get(clave, valor_por_defecto): obtener valor sin error si la clave no existe
print("\nUso de get():")
print(estudiante.get('correo', 'Correo no registrado'))  # Devuelve el valor por defecto

# 5. dict.update(otro_diccionario): agrega o actualiza elementos
print("\nUsamos update() para agregar un nuevo dato:")
estudiante.update({'correo': 'lucia@example.com'})
print(estudiante)

# 6. dict.pop(clave): elimina una clave específica y devuelve su valor
print("\nEliminamos la clave 'edad' con pop():")
edad_eliminada = estudiante.pop('edad')
print("Edad eliminada:", edad_eliminada)
print("Diccionario actualizado:", estudiante)


# 7. dict.clear(): elimina todos los elementos, pero no elimina el diccionario en sí
print("\nUsamos clear() para borrar todo el diccionario:")
estudiante.clear()
print("Diccionario vacío:", estudiante)  # {}
```

**EJEMPLO. Agenda de contactos**

Crea una función que permita agregar contactos a una agenda usando un diccionario. Cada contacto tendrá un nombre como clave y un número de teléfono como valor. La función debe permitir buscar un contacto por nombre.

```python
# Diccionario principal donde se almacenan los contactos
agenda = {}

# Función para agregar un contacto
def agregar_contacto(nombre, telefono):
    if nombre in agenda:
        print(f"El contacto '{nombre}' ya existe. Se actualizará el número.")
    agenda[nombre] = telefono
    print(f"Contacto '{nombre}' agregado o actualizado correctamente.")

# Función para buscar un contacto
def buscar_contacto(nombre):
    if nombre in agenda:
        print(f"{nombre}: {agenda[nombre]}")
    else:
        print(f"El contacto '{nombre}' no está en la agenda.")

# Programa principal
try:
    print("Agregar primer contacto:")
    nombre1 = input("Ingrese el nombre: ").strip().capitalize()
    telefono1 = input("Ingrese el teléfono: ").strip()

    # Validación: verificar que no estén vacíos
    if not nombre1 or not telefono1:
        raise ValueError("Debe ingresar tanto el nombre como el teléfono del primer contacto.")

    agregar_contacto(nombre1, telefono1)

    print("\nAgregar segundo contacto:")
    nombre2 = input("Ingrese el nombre: ").strip().capitalize()
    telefono2 = input("Ingrese el teléfono: ").strip()

    if not nombre2 or not telefono2:
        raise ValueError("Debe ingresar tanto el nombre como el teléfono del segundo contacto.")

    agregar_contacto(nombre2, telefono2)

    # Búsqueda de contactos
    print("\nBuscar contactos:")

    nombre_busqueda1 = input("¿A quién desea buscar?: ").strip().capitalize()
    if not nombre_busqueda1:
        raise ValueError("Debe ingresar un nombre para buscar el primer contacto.")
    buscar_contacto(nombre_busqueda1)

    nombre_busqueda2 = input("¿Desea buscar otro contacto? Ingrese la información: ").strip().capitalize()
    if not nombre_busqueda2:
        raise ValueError("Debe ingresar un nombre para buscar el segundo contacto.")
    buscar_contacto(nombre_busqueda2)

except ValueError as ve:
    print("Error:", ve)

except Exception as e:
    print("Ocurrió un error inesperado:", e)
```

Errores de varios tipos:
ValueError (convesión)
TypeError (operaciones incompatibles)
IndexError (acceso a un índice fuera de rango)
ZeroDivisionError (división por cero)
KeyError (Acceso a una clave que no existe en un diccionario)
NameError (Uso de una variable no definida)
AttributeError (Llamada a un método que no existe)

**TUPLAS**

Son estructuras similares a las listas, ya que permiten guardar varios elementos en una sola variable, pero con la diferencia de que no se pueden modificar una vez creadas.

Características

*   Se declara entre paréntesis ().
*   Son inmutables (no se puede cambiar, agregar ni eliminar elementos una vez creada).
*   Pueden contener datos de diferentes tipos (números, textos, booleanos, otras tuplas)
*   útiles cuando no se requiere modificar los datos.
*  Se accede a los elementos con índices, igual que en las listas.

Estructura General

tupla = (elemento1, elemento2, elemento3, ..., elementoN)

**EJEMPLO. INFORMACIÓN GEOGRÁFICA**

Define una función que reciba una tupla con la siguiente estructura: (país, capital, población). La función debe mostrar la información y decir si la población supera los 10 millones de habitantes.

```python
# Definición de la función
def analizar_pais(datos_pais):
    pais, capital, poblacion = datos_pais  # Desempaquetado directo
    print(f"\n Análisis del país:")
    print(f"País: {pais}")
    print(f"Capital: {capital}")
    print(f"Población: {poblacion} millones")

    if poblacion >= 20:
        print("Este país tiene una población alta.")
    elif 10 <= poblacion < 20:
        print("Este país tiene una población media.")
    else:
        print("Este país tiene una población baja.")

#Programa Principal

try:
    pais = input("Ingrese el nombre del país: ").strip().capitalize()
    capital = input("Ingrese la capital del país: ").strip().capitalize()
    #Replace (reemplaza  todas las comas por puntos, para convertir de manera correcta a float)
    poblacion_str = input("Ingrese la población (en millones): ").replace(",", ".").strip()

    # Validar campos vacíos antes de convertir
    if not pais or not capital or not poblacion_str:
        raise Exception("Debe ingresar datos en los tres campos.")

    # Intentar conversión a float
    poblacion = float(poblacion_str)

    # Validar que no sea negativa
    if poblacion < 0:
        raise ValueError("La población no puede ser negativa.")

    # Empaquetar tupla y llamar a la función
    datos_pais = (pais, capital, poblacion)
    analizar_pais(datos_pais)

#Dar mensajes específicos, según el tipo de error
except ValueError:
    print("Error: La población debe ser un número válido.")
except Exception as e:
    print("Error inesperado:", e)
```

**SETS (CONJUNTOS)**

Los sets son especialmente útiles cuando se necesita trabajar con colecciones sin elementos repetidos y aplicar operaciones como unión, intersección o diferencia entre conjuntos. También permiten eliminar duplicados de forma eficiente.

Características
*   Se declaran con llaves {} o la función set().
*   No permiten duplicados.
*   Son no ordenados.
*   Son mutables, pero no permiten acceso por índice.

Estructura General

mi_set = {elemento1, elemento2, elemento3}




**MÉTODOS**

Los métodos más usados en los sets, son:
1. add(elem):Agrega un nuevo elemento al set,mi_set.add(5)
2. remove(elem):Elimina un elemento (lanza error si no existe), mi_set.remove(3)
3. discard(elem):	Elimina un elemento si existe (no lanza error si no está),	mi_set.discard(4)
4. pop():	Elimina un elemento aleatorio del set,	mi_set.pop()
5. clear():Elimina todos los elementos del set,	mi_set.clear()
6. copy():	Devuelve una copia del set,	nuevo = mi_set.copy()
7. update(iterable):Agrega varios elementos desde otra        colección,	mi_set.update([6, 7, 8])

**CREAR UN SETS (CONJUNTO)**

**ELIMINAR UN ELEMENTO DE UN SET**

Los métodos discard() o remove(), permiten eliminar un elemento del set. La diferencia radica en que, discard() no genera un error si el elemento no está presente en el conjunto, mientras que,remove(),si.

```python
#Usando remove
set_2 = {5, 4, 6, 8, 8, 1} #Se puede crear usando solo las llaves
set_2.remove(8)
print(set_2)

#Usando discard
set_2 = {5, 4, 6, 8, 8, 1} #Se puede crear usando solo las llaves
set_2.discard(5)
print(set_2)

#Usando pop
set_2 = {5, 4, 6, 8, 8, 1} #Se puede crear usando solo las llaves
set_2.pop()
print(set_2)

#Usando clear
set_2 = {5, 4, 6, 8, 8, 1} #Se puede crear usando solo las llaves
set_2.clear()
print(set_2)
```

**OPERACIONES DE CONJUNTOS CON SETS**
1. union() o: Une dos sets (sin duplicados).
2. intersection() o &: Elementos en común.
3. difference() o -: Elementos en a que no están en b.
4. symmetric_difference() o ^: Elementos que están en uno, pero no en ambos.
5. issubset():¿Todos los elementos de a están en b?
6. issuperset():¿a contiene todos los elementos de b?
7. isdisjoint():¿No comparten ningún elemento?

**EJEMPLO**

```python
# Crear dos sets
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Métodos
a.add(7)
a.discard(2)

# Operaciones entre conjuntos
print("Set A:", a)                             # {1, 3, 4, 7}
print("Set B:", b)                             # {3, 4, 5, 6}

print("Unión:", a | b)                         # {1, 3, 4, 5, 6, 7}
print("Intersección:", a & b)                  # {3, 4}
print("Diferencia A - B:", a - b)              # {1, 7}
print("Diferencia B - A:", b - a)              # {5, 6}
print("Diferencia simétrica:", a ^ b)          # {1, 5, 6, 7}

# Más operaciones lógicas
print("¿A es subconjunto de B?", a.issubset(b))        # False ¿Todos los elementos de A están en B?
print("¿B es superconjunto de A?", b.issuperset(a))    # False ¿B contiene todos los elementos de A?
print("¿A y B no tienen elementos en común?", a.isdisjoint(b))  # False ¿A y B no comparten ningún elemento?
```

**EJEMPLO: INTERESES COMUNES**

Crea una función que reciba dos conjuntos de intereses de dos personas y devuelva los intereses en común, los que son únicos para cada persona, y la unión total de intereses.

```python
#Definición de función
import sys
def comparar_intereses(set1, set2):
    """Compara dos sets de intereses y muestra los resultados."""
    comunes = set1 & set2
    unicos1 = set1 - set2
    unicos2 = set2 - set1
    todos = set1 | set2

    print("\nResultado de la comparación:")
    print(f"Intereses comunes       : {comunes}")
    print(f"Solo en Persona 1       : {unicos1}")
    print(f"Solo en Persona 2       : {unicos2}")
    print(f"Todos los intereses     : {todos}")

#Programa principal
try:
    # Ingresar intereses como cadenas separadas por coma
    intereses_1 = input("Ingrese los intereses de la Persona 1 (separados por coma): ").strip()
    intereses_2 = input("Ingrese los intereses de la Persona 2 (separados por coma): ").strip()

    # Validar que no estén vacíos

    if not intereses_1 or not intereses_2:
      #Nos informa que hay un problema
      #Capturar el error para hacer algo (mostrarlo al usuario o registrarlo en un sistema)
        raise ValueError("Ambas personas deben tener al menos un interés.")


    #Convertir las entradas a sets
    set1 = set(intereses_1.lower().split(","))
    set2 = set(intereses_2.lower().split(","))

    # Eliminar espacios en blanco extra
    set1 = {interes.strip() for interes in set1}
    set2 = {interes.strip() for interes in set2}

    # Llamar a la función
    comparar_intereses(set1, set2)

except Exception as e:
    print("Ocurrió un error inesperado:", e)
```
