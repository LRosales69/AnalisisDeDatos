EJERCICIO 1.

```python
def verificar_paciente(paciente):
    if paciente["edad"] >= 18:
        print(f"{paciente['nombre']} es mayor de edad.")
    else:
        print(f"{paciente['nombre']} es menor de edad.")
    print("Edad:", paciente["edad"])


# Programa principal
try:
    nombre = input("Ingrese el nombre del paciente: ").strip()
    edad_str = input("Ingrese la edad: ").strip()
    peso_str = input("Ingrese el peso en kg: ").strip()

    if not nombre or not edad_str or not peso_str:
        raise ValueError("Todos los campos son obligatorios.")

    edad = int(edad_str)
    peso = float(peso_str)

    if edad < 0 or peso <= 0:
        raise ValueError("Edad o peso inválido.")

    datos = {"nombre": nombre, "edad": edad, "peso": peso}
    verificar_paciente(datos)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Error inesperado:", e)
```

EJERCICIO 2.

```python
def agregar_salario(perfil, salario):
    perfil["salario"] = salario
    print("Salario registrado:", perfil["salario"])


# Programa principal
try:
    nombre = input("Ingrese el nombre: ").strip()
    cargo = input("Ingrese el cargo: ").strip()
    salario_str = input("Ingrese el salario mensual: ").strip()

    if not nombre or not cargo or not salario_str:
        raise ValueError("Debe completar todos los campos.")

    salario = float(salario_str)
    if salario <= 0:
        raise ValueError("El salario debe ser mayor que 0.")

    perfil = {"nombre": nombre, "cargo": cargo}
    agregar_salario(perfil, salario)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Error inesperado:", e)
```

```python
ef verificar_rol(usuario):
    if usuario["rol"].lower() == "admin":
        print("Acceso total")
    else:
        print("Acceso limitado")


# Programa principal
try:
    nombre = input("Ingrese el nombre del usuario: ").strip()
    rol = input("Ingrese el rol del usuario (admin/otro): ").strip()

    if not nombre or not rol:
        raise ValueError("Todos los campos son obligatorios.")

    usuario = {"nombre": nombre, "rol": rol}
    verificar_rol(usuario)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Error inesperado:", e)
```

```python
def nota_matematicas(notas):
    print("Nota de matemáticas:", notas["matemáticas"])


# Programa principal
try:
    mate_str = input("Ingrese la nota de matemáticas: ").strip()
    lengua_str = input("Ingrese la nota de lengua: ").strip()

    if not mate_str or not lengua_str:
        raise ValueError("Ambas notas son requeridas.")

    mate = float(mate_str)
    lengua = float(lengua_str)

    if not (0 <= mate <= 10 and 0 <= lengua <= 10):
        raise ValueError("Las notas deben estar entre 0 y 10.")

    notas = {"matemáticas": mate, "lengua": lengua}
    nota_matematicas(notas)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Error inesperado:", e)
```

```python
def evaluar_bono(empleado):
    if empleado["desempeño"].lower() == "excelente":
        print("Empleado recibe bono")
    else:
        print("No califica para bono")


# Programa principal
try:
    nombre = input("Ingrese el nombre del empleado: ").strip()
    desempeño = input("Ingrese el nivel de desempeño (excelente, bueno, regular): ").strip()

    if not nombre or not desempeño:
        raise ValueError("Debe ingresar el nombre y desempeño.")

    empleado = {"nombre": nombre, "desempeño": desempeño}
    evaluar_bono(empleado)

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Error inesperado:", e)
```
