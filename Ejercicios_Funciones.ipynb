```python
#Ejercicio 1.
# Definición de la función que calcula la comisión
def calcular_comision(ventas):
      if ventas > 1000000:
        comision = ventas * 0.03
      else:
        comision = ventas * 0.01
      return comision

# Programa principal
try:
    # Solicitar al usuario el total de ventas
    total_ventas = float(input("Ingrese el total de ventas en pesos chilenos (CLP): "))

    # Validar que las ventas no sean negativas
    if total_ventas < 0:
        print("Error: El monto de ventas no puede ser negativo.")
    else:
        # Calcular la comisión usando la función
        comision_vendedor = calcular_comision(total_ventas)

        # Mostrar el resultado con formato monetario
        print(f"La comisión que corresponde es: ${comision_vendedor:,.0f} CLP")

except ValueError:
    print("Error: Debe ingresar un número válido para las ventas.")
```

```python
#Ejercicio 2.

def calcular_costo_llamada(duracion_minutos):
    if duracion_minutos <= 3:
        costo = 400
    else:
        minutos_adicionales = duracion_minutos - 3
        costo = 400 + (minutos_adicionales * 150)
    return costo

# Programa principal
try:
    # Solicitar al usuario la duración de la llamada en minutos
    duracion = int(input("Ingrese la duración de la llamada en minutos: "))

    # Validar que la duración sea un número positivo
    if duracion < 0:
        print("Error: La duración no puede ser negativa.")
    else:
        # Calcular el costo usando la función
        costo_total = calcular_costo_llamada(duracion)

        # Mostrar el resultado
        print(f"El costo total de la llamada es: ${costo_total} CLP")

except ValueError:
    print("Error: Debe ingresar un número entero válido para la duración.")
```

```python
#Ejercicio 3.
## Definición de la función que determina el mayor de tres números
def mayor_de_tres(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

# Programa principal
try:
    # Solicitar al usuario que ingrese tres números
    numero1 = float(input("Ingresa el primer número: "))
    numero2 = float(input("Ingresa el segundo número: "))
    numero3 = float(input("Ingresa el tercer número: "))

    # Llamar a la función para determinar el mayor
    mayor = mayor_de_tres(numero1, numero2, numero3)

    # Mostrar el resultado
    print(f"El número mayor es: {mayor}")

except ValueError:
    print("Error: Debes ingresar valores numéricos válidos.")
```

```python
#Ejercicio 4.
# Definición de la función
def clasificar_numero(x):

    if 0 <= x < 5:
        return "El número es menor que 5."
    elif 5 <= x <= 50:
        return "El número está entre 5 y 50 (inclusive)."
    else:  # x > 50
        return "El número es mayor que 50."

# Programa principal
try:
    # Solicitar al usuario ingresar un número entero
    numero = int(input("Ingresa un número entero: "))
    if numero < 0:
       print("El número es negativo.")
    else:
    # Llamar a la función para clasificar el número
       mensaje = clasificar_numero(numero)
    # Mostrar el resultado
       print(mensaje)
except ValueError:
    print("Error: Debes ingresar un número entero válido.")
```

```python
#Ejercicio 5.
#Definición de la función
def clasificar_categoria(edad):
    if edad < 12:
        return "Infantil"
    elif 12 <= edad <= 17:
        return "Juvenil"
    elif 18 <= edad <= 40:
        return "Adulto"
    else:  # edad > 40
        return "Senior"

# Programa principal
try:
    edad_usuario = int(input("Ingresa tu edad: "))
    if edad_usuario < 0:
        print("Error: La edad no puede ser negativa.")
    else:
        categoria = clasificar_categoria(edad_usuario)
        print(f"Tu categoría deportiva es: {categoria}")
except ValueError:
    print("Error: Debes ingresar un número entero válido para la edad.")
```

```python
#Ejercicio 6.
#Definición de la función
def clasificar_imc(peso, altura):
    imc = peso / (altura ** 2)
    if imc < 18.5:
        return f"IMC: {round(imc, 2)} - Bajo peso"
    elif 18.5 <= imc < 25:
        return f"IMC: {round(imc, 2)} - Peso normal"
    elif 25 <= imc < 30:
        return f"IMC: {round(imc, 2)} - Sobrepeso"
    else:
        return f"IMC: {round(imc, 2)} - Obesidad"

# Programa principal
try:
    peso = float(input("Ingresa tu peso en kg: "))
    altura = float(input("Ingresa tu altura en metros: "))

    if peso <= 0 or altura <= 0:
        print("Error: El peso y la altura deben ser valores positivos.")
    else:
        resultado = clasificar_imc(peso, altura)
        print(resultado)

except ValueError:
    print("Error: Debes ingresar valores numéricos válidos.")
```
