```python
#Ejercicio 1. Semáforo
semaforo = input("Color del semáforo (rojo, amarillo, verde): ").lower()

if semaforo == "verde":
    print("Cruzar la calle")
elif semaforo == "amarillo":
    print("Prepararse para detenerse")
elif semaforo == "rojo":
    print("Esperar")
else:
    print("Color no reconocido")
```

```python
#Ejercicio 2. Verificar mayoría de edad

edad = int(input("¿Cuántos años tienes? "))

if edad < 0:
    print("Edad inválida. No puedes tener una edad negativa.")
elif edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

```python
#Ejercicio 3: Número positivo, cero o negativo
numero = float(input("Ingresa un número: "))

if numero > 0:
    print("El número es positivo.")
elif numero == 0:
    print("El número es cero.")
else:
    print("El número es negativo.")
```

```python
#Ejercicio 4. Si gasto hasta $100, pago con dinero en efectivo. Sino, si gasto más de $100 pero menos de $300, pago con tarjeta de débito.
#Sino, pago con tarjeta de crédito.

gasto = float(input("¿Cuánto gastaste en tu compra? "))
if gasto <= 100:
    print("Pago con dinero en efectivo.")
elif gasto > 100 and gasto < 300:
    print("Pago con tarjeta de débito.")
else:
    print("Pago con tarjeta de crédito.")
```

```python
#Ejercicio 5. Compra
totalcompra = float(input("Ingresa el total de la compra: "))

if totalcompra > 200:
    tasadescuento = 20
    importedescuento = totalcompra * tasadescuento / 100
    importeapagar = totalcompra - importedescuento
    print("Se aplica un descuento del 20%.")
elif totalcompra > 100:
    tasadescuento = 10
    importedescuento = totalcompra * tasadescuento / 100
    importeapagar = totalcompra - importedescuento
    print("Se aplica un descuento del 10%.")
else:
    importeapagar = totalcompra
    print("No aplica descuento.")

print("El importe a pagar es:", importeapagar)
```
